## Hot & Cold Learning ðŸ¥µ/ðŸ¥¶

```python
## page-54

import matplotlib.pyplot as plt

weight = 0.5
input_ = 0.5
goal_pred = 0.8

step_amount = 10
error_list = []
pred_list = []
num_iters = 100
for iter_ in range(num_iters):
    pred = input_ * weight
    error = (pred - goal_pred) ** 2
    ##print(f"Error={error} \t Pred={pred}")
    
    up_pred = input_ * (weight + step_amount)
    up_error = (goal_pred - up_pred) ** 2
    
    down_pred = input_ * (weight - step_amount)
    down_error = (goal_pred - down_pred) ** 2
    
    if down_error < up_error:
        weight -= step_amount             ## go down. decrease the weight
        error_list.append(down_error)
        pred_list.append(down_pred)
    elif up_error <= down_error:
        weight += step_amount             ## go up. increase the weight
        error_list.append(up_error)
        pred_list.append(up_pred)
        
plt.plot(list(range(num_iters)), error_list, label="error")
plt.plot(list(range(num_iters)), pred_list, label="pred")
plt.legend(); plt.grid(); plt.tight_layout(); plt.show()
```
