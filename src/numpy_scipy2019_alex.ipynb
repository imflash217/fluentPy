# Using `.flatten()` and `.ravel()`

### `.flatten()` return a COPY
### `.ravel()` returns a VIEW (if possible i.e. the data is _contiguous_; else returns a COPY)

````python
a = np.array([[1,2],[11,12]])
a
````

````python
a.shape
````

````python
f = a.flatten()
f
````

````python
f = np.flatten(a)
f
````

````python
a, f
````

````python
a[0,0] = 1000                      ## changing a does NOT change f.......so .flatten() return a COPY
a, f 
````

````python
a = np.array([[0,1],[2,3]])
a
````

````python
r = a.ravel()
r
````

````python
a, r
````

````python
a[0, 0] = 999                 ## changing a changes the r. so .ravel() returns a VIEW (if possible i.e. if the data is contiguous)
a, r
````

````python
r[3] = 7777777777777          ## changing a changes the r. so .ravel() returns a VIEW (if possible i.e. if the data is contiguous)
a, r
````
